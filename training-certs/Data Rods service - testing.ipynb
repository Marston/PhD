```python
#%%time
import os
import netrc
import requests
from requests.auth import HTTPBasicAuth

# get to the token page. This is a redirect
# handling secrets
tokenURL = "https://s6qtl5s0b4.execute-api.us-east-1.amazonaws.com/UAT/signin"
baseURI = "https://qx0e40pab7.execute-api.us-east-1.amazonaws.com/UAT/timeseries"

os.environ['TOKEN'] = requests.get(tokenURL, auth=HTTPBasicAuth(netrc.netrc().hosts['uat.urs.earthdata.nasa.gov'][0], 
                                                                netrc.netrc().hosts['uat.urs.earthdata.nasa.gov'][2]),
                                   allow_redirects=True).text.replace('"','')
lat = 42.875
lon = 15.625
time_start = "1999-10-30T00:00:00" 
time_end = "1999-11-02T23:59:59"
#variable = "GLDAS_NOAH025_3H_2_1_Tair_f_inst"
variable = "GPM_3IMERGHH_06_precipitationCal"

payload = {"data": variable,
           "location": "[{},{}]".format(lat,lon),
           "time": "{}/{}".format(time_start,time_end)}

headers = {"authorizationtoken": os.environ.get('TOKEN')}
response = requests.get(baseURI, params=payload, headers=headers)

print(response.text)
           
if os.environ['TOKEN']: del os.environ['TOKEN']
```

```python
# split the content on '\n' so that you can view it as rows
# print the output as rows and decode the binary content to normal ascii string
# I've also added some line numbers based on the assumption that the output has less
# than 1000 lines
for lineNum, line in enumerate(response.content.splitlines(), start=0): 
    print(str(lineNum).zfill(6),'\t',line.decode('ascii'))
```
